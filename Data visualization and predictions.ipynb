```python
import bs4 as bs
import pickle
import requests

def save_sp500_tickers():
    resp = requests.get('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
    soup = bs.BeautifulSoup(resp.text)
    table = soup.find('table', {'id': 'constituents'})
    tickers = []
    for row in table.findAll('tr')[1:]:
        ticker = row.find('td').text
        tickers.append(ticker)

    with open("sp500tickers.pickle", "wb") as f:
        pickle.dump(tickers, f)

    print(tickers)
    print('hi')

    
    return tickers
save_sp500_tickers()
```

```python
import pandas as pd
import yfinance as yf
import mplfinance as mpf
import os
import pickle

def get_sp500_tickers():

    sp_wiki_url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
    sp_wiki_df_list = pd.read_html(sp_wiki_url)
    sp_df = sp_wiki_df_list[0]
    sp_ticker_list = list(sp_df['Symbol'].values)

    with open('sp500_tickers.pkl', 'wb') as file:
        pickle.dump(sp_ticker_list, file)
```

```python
def compile_finances():

    with open('sp500_tickers.pkl', 'rb') as file: 
      
        sp_ticker_list = pickle.load(file) 
    
    if not os.path.exists('stock_dfs'):
        os.makedirs('stock_dfs')
        
    for ticker in sp_ticker_list[:3]:

        if not os.path.exists('stock_dfs/{}.csv'.format(ticker)):

            stock_df = yf.Ticker(ticker)
            pnl = stock_df.financials
            bs = stock_df.balancesheet
            cf = stock_df.cashflow
            fs = pd.concat([pnl, bs, cf])

            fs.to_csv('stock_dfs/{}.csv'.format(ticker))

        else:
            print('No current need for {}'.format(ticker))
```

```python
def compile_stock_data():

    with open('sp500_tickers.pkl', 'rb') as file: 
      
        sp_ticker_list = pickle.load(file)
        
    if not os.path.exists('market_dfs'):
        os.makedirs('market_dfs')
        
    for ticker in sp_ticker_list[:3]:

        if not os.path.exists('market_dfs/{}.csv'.format(ticker)):

            stock_df = yf.download(ticker, start='2020-1-1')
            stock_df.to_csv('market_dfs/{}.csv'.format(ticker))
        
        else:
            
            print('No current need for {}'.format(ticker))
```

```python
def compile_data(): 
    
    with open('sp500_tickers.pkl', 'rb') as file: 
      
        sp_ticker_list = pickle.load(file)

    main_df = pd.DataFrame()

    for count, ticker in enumerate(sp_ticker_list[:3]):
        df = pd.read_csv('market_dfs/{}.csv'.format(ticker))
        df.set_index('Date', inplace=True)

        df.rename(columns = {'Adj Close': ticker}, inplace=True)
        df.drop(['Open','High','Low','Close','Volume'], axis=1, inplace=True)

        if main_df.empty:
            main_df = df
            
        else:
            main_df = main_df.join(df, how='outer')

    print(main_df.head())
    main_df.to_csv('Combined_AdjClose')

compile_data()
```

```python
from matplotlib import style
import matplotlib.pyplot as plt
import numpy as np

style.use('ggplot')

def visualize_data():

    df = pd.read_csv('Combined_AdjClose')
    df.set_index('Date', inplace=True)
    df_corr = df.pct_change().corr()

    data = df_corr.values
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    
    heatmap = ax.pcolor(data, cmap=plt.cm.RdYlGn)
    fig.colorbar(heatmap)
    ax.set_xticks(np.arange(data.shape[0]) + 0.5, minor=False)
    ax.set_yticks(np.arange(data.shape[1]) + 0.5, minor=False)
    ax.invert_yaxis()
    ax.xaxis.tick_top()

    column_labels = df_corr.columns
    row_labels = df_corr.index

    ax.set_xticklabels(column_labels)
    ax.set_yticklabels(row_labels)
    plt.xticks(rotation=90)
    heatmap.set_clim(-1, 1)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd

def process_data_for_labels(ticker):

    hm_days = 7
    df = pd.read_csv('Combined_AdjClose', index_col=0)
    tickers = df.columns.values.tolist()
    df.fillna(0, inplace=True)

    for i in range(1, hm_days+1):
        df['{}_{}d'.format(ticker, i)] = (df[ticker].shift(-i) - df[ticker]) / df[ticker]
        
    df.fillna(0, inplace=True)

    return tickers, df

process_data_for_labels('MMM')
```

```python
def buy_sell_hold(cols):

    requirement = 0.02

    for col in cols:
        
        if col > requirement:
            return 1
        if col < -requirement:
            return -1

    return 0
```

```python
from collections import Counter

def extract_featuresets(ticker):

    tickers, df = process_data_for_labels(ticker)

    df['{}_target'.format(ticker)] =  list(map(buy_sell_hold, df[[c for c in df.columns if c not in tickers]].values))
    print(df)
    vals = df['{}_target'.format(ticker)].values.tolist()
    str_vals = [str(i) for i in vals]
    print('Data spread:', Counter(str_vals))

    df.fillna(0, inplace=True)
    df = df.replace([np.inf, -np.inf], np.nan)
    df.dropna(inplace=True)

    df_vals = df[[ticker for ticker in tickers]].pct_change()
    df_vals = df_vals.replace([np.inf, -np.inf], np.nan)
    df_vals.fillna(0, inplace=True)

    X = df_vals.values
    y = df['{}_target'.format(ticker)].values

    return X, y, df

extract_featuresets('MMM')

```

```python
from sklearn import svm, model_selection, neighbors
from sklearn.ensemble import VotingClassifier, RandomForestClassifier

def do_ml(ticker):

    X, y, df = extract_featuresets(ticker)
    X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size = 0.25)

    clf = VotingClassifier([('lsvc', svm.LinearSVC(dual=True)), 
                            ('knn', neighbors.KNeighborsClassifier()),
                            ('rfor', RandomForestClassifier())])
    
    clf.fit(X_train, y_train)
    confidence = clf.score(X_test, y_test)

    predictions = clf.predict(X_test)
    print('Prediction spread:', Counter(predictions))

    return confidence

do_ml('MMM')

```
